```python
import pandas as pd
import numpy as np
```

```python
companies = pd.read_excel("companies.xlsx")
balancesheet = pd.read_excel("balancesheet.xlsx")
cashflow = pd.read_excel("cashflow.xlsx")
profitandloss = pd.read_excel("profitandloss.xlsx")
analysis = pd.read_excel("analysis.xlsx")
prosandcons = pd.read_excel("prosandcons.xlsx")
documents = pd.read_excel("documents.xlsx")
```

```python
companies.head()
```

```python
def fix_headers(df):
    df.columns = df.iloc[0]   # Set first row as header
    df = df[1:]               # Remove first row
    df.reset_index(drop=True, inplace=True)
    return df
```

```python
companies = fix_headers(companies)
balancesheet = fix_headers(balancesheet)
cashflow = fix_headers(cashflow)
profitandloss = fix_headers(profitandloss)
analysis = fix_headers(analysis)
prosandcons = fix_headers(prosandcons)
documents = fix_headers(documents)
```

```python
profit_columns = [
    'sales',
    'expenses',
    'operating_profit',
    'opm_percentage',
    'other_income',
    'interest',
    'depreciation',
    'profit_before_tax',
    'tax_percentage',
    'net_profit',
    'eps',
    'dividend_payout'
]

for col in profit_columns:
    profitandloss[col] = pd.to_numeric(
        profitandloss[col],
        errors='coerce'
    )
```

```python
profitandloss.info()
```

```python
balancesheet.info()
```

```python
balance_columns = [
    'equity_capital',
    'reserves',
    'borrowings',
    'other_liabilities',
    'total_liabilities',
    'fixed_assets',
    'cwip',
    'investments',
    'other_asset',
    'total_assets'
]

for col in balance_columns:
    balancesheet[col] = pd.to_numeric(
        balancesheet[col],
        errors='coerce'
    )
```

```python
cashflow_columns = [
    'operating_activity',
    'investing_activity',
    'financing_activity',
    'net_cash_flow'
]

for col in cashflow_columns:
    cashflow[col] = pd.to_numeric(
        cashflow[col],
        errors='coerce'
    )
```

```python
profitandloss['year'].unique()
```

```python
profitandloss['fiscal_year'] = profitandloss['year'].str.extract(r'(\d{4})')

profitandloss['fiscal_year'] = pd.to_numeric(
    profitandloss['fiscal_year'],
    errors='coerce'
)
```

```python
profitandloss['year'].unique()
```

```python
balancesheet['fiscal_year'] = balancesheet['year'].str.extract(r'(\d{4})')

balancesheet['fiscal_year'] = pd.to_numeric(
    balancesheet['fiscal_year'],
    errors='coerce'
)
```

```python
balancesheet['year'].unique()
```

```python
cashflow['fiscal_year'] = cashflow['year'].str.extract(r'(\d{4})')

cashflow['fiscal_year'] = pd.to_numeric(
    cashflow['fiscal_year'],
    errors='coerce'
)
```

```python
cashflow['year'].unique()
```

```python
balancesheet['debt_to_equity'] = (
    balancesheet['borrowings'] /
    (
        balancesheet['equity_capital'] +
        balancesheet['reserves']
    )
)
```

```python
profitandloss['net_profit_margin_pct'] = (
    profitandloss['net_profit'] /
    profitandloss['sales']
) * 100
```

```python
profitandloss['interest_coverage'] = (
    profitandloss['operating_profit'] /
    profitandloss['interest']
)
```

```python
cashflow['free_cash_flow'] = (
    cashflow['operating_activity'] +
    cashflow['investing_activity']
)
```
