```python
import pandas as pd

profitandloss = pd.read_csv("eda_profitandloss.csv")
balancesheet = pd.read_csv("eda_balancesheet.csv")
cashflow = pd.read_csv("eda_cashflow.csv")
```

```python
profitandloss.columns
```

```python
profitandloss.describe()
```

```python
import matplotlib.pyplot as plt

sales_trend = profitandloss.groupby("fiscal_year")["sales"].sum()

sales_trend.plot(marker="o")
plt.title("Total Sales Over Years")
plt.show()
```

```python
profit_trend = profitandloss.groupby("fiscal_year")["net_profit"].sum()

profit_trend.plot(figsize=(10,5))
```

```python
top_sales = profitandloss.groupby("company_id")["sales"].sum().sort_values(ascending=False).head(10)

top_sales.plot(kind="bar")
```

```python
top_profit = profitandloss.groupby("company_id")["net_profit"].sum().sort_values(ascending=False).head(10)

top_profit.plot(kind="bar")
```
